# 11章 カテゴリ
## 11-1 カテゴリ型への変換
### Q: 性別をカテゴリに適したデータ型に変換
#### Awesome

```python
%%bigquery
select
    sex,
    case
        when sex = "M" then 1
        when sex = "F" then 2
        when sex = "NA" then 3
    end as sex_cat
from example.customer
```

## 11-3 カテゴリ値の集約
### Q: 年齢カテゴリの集約
#### Awesome

```python
%%bigquery
select
    * except (age),
    case when age < 60 then floor(age / 10) else 6 end as age_cat
from example.customer
```

## 11-4 カテゴリ値の組み合わせ
### Q: 性別と年齢のカテゴリの組み合わせ
#### Awesome

```python
%%bigquery
select
    *,
    concat(
        coalesce(sex, "unknown"),
        "_",
        floor(age / 10)
    ) as sex_age_cat
from example.customer
```
